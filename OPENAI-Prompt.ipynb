```python
# Configura tu clave de API
openai.api_key = "secret-key"

context = """ 
1. Teamwork Thinkers:
Participates in interdisciplinary teams, seeking and offering collaboration and support outside their area.
Adjusts their communication style taking into account the receptor’s needs and the situation at hand.
Answers for technical inquiries their team and project have and escalates if necessary.
Challenges ideas to the team, takes an active part in group discussions, taking into account others’ opinions.

2. Raising Leaders:
Delegates tasks and work processes to team members while assuming responsibility for results.
Identifies development needs of your work team and jointly defines improvement plans.
Provides feedback on the management of its collaborators in a respectful and cordial environment, based on the responsibilities of the roles and competencies of its team.
Take the initiative and participate proactively in projects by sharing different points of view and visualizing risks in advance.

3. Business Booster:
Enables new processes and changes based on the market dynamics and can transmit the sense of urgency to the team.
Ensures workforce development based on the local strategy.
Creates and proposes preventive plans to the team and management to avoid an issue from becoming recurrent.
Generates ideas and obtains positive results in the processes under their responsibility building unique perspectives, knowledges and skills.

4. Partner Builders:
Initiates conversations with the client to attend and clarify concerns/inquiries. Manage the conversation effectively.
It attends to each client with dedication and willingness to meet the demands that arise, demonstrating transversal knowledge of PGD.
Provides feedback back to the team in a respectful and professional manner to bring clarity to internal and external client concerns/inquiries.
Demonstrates knowledge, passion and professionalism to clients, becoming a visible SME to the client and capability.

5. Innovation Seekers:
Encourages creativity on the part of others by highlighting the benefits of innovation and change on the performance of a specific area.
Substantially improves work processes through the early adoption of innovative approaches or technologies within and outside of their teams.
Facilitates spaces where creativity and innovation can be fostered, to implement new ideas to solve problems and improve processes.
Continuously presents solutions and ideas that solve present or future problems or situations within or impacting the client, applying new knowledge."""

question = """With this feedback: Steve is always available, trying to help as much as possible but questioning much of the steps we have defined, which is something good and always try to help his partners to succeed as as well. One area of improvement could be improve his presentation skills and finding new ways to improve certain processes. 
/n What is the related competence? and what is the justification?"""

# Combina el contexto y la pregunta en un solo prompt
prompt = f"Context: {context}\nQuestion: {question}"

# Llama a la función de OpenAI para obtener una respuesta
response = openai.Completion.create(
    engine="text-davinci-003",  # Utiliza el motor GPT-3.5 Turbo
    prompt=prompt,
    max_tokens=100  # Puedes ajustar este valor según la longitud de la respuesta deseada
)

# Imprime la respuesta
print(response.choices[0].text.strip())
```
